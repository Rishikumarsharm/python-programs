```python
##bar plot
x=["mp","up","rj","mh"]
y=[1000,2000,3000,4000]
colours=["red","green","skyblue","yellow"]
hatch_pattern=["////","\\\\","--","o"]
plt.bar(x,y,color=colours,width=0.5,edgecolor="black",hatch=hatch_pattern,label=x)
plt.legend()
plt.grid()
```

```python
##histrogram --> it use in distribution of data in bins(intervals)
data=np.random.rand(1000)
plt.hist(data,bins=30)
plt.show()
```

```python
#piechart == distribution of data in slices
labels=["a","b","c","d"]
sizes=[10,20,30,40]
explodes=[0.3,0.5,0.6,0.1]
plt.pie(sizes,labels=labels,autopct="%1.2f%%",explode=explodes,shadow=True)
plt.title("               pie chart")
plt.show()
```

```python
##subplot --multiple plot in a single figure
x=np.linspace(0,10,100)
plt.figure(figsize=(10,5))
plt.subplot(2,2,1) #2 rows ,2 columns, 1st subplot
plt.plot(x,np.sin(x))
plt.title("sin(x)")
```

```python
x=np.linspace(0,10,100)
plt.figure(figsize=(10,5))
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.title("cos(x)")
```

```python
x=np.linspace(0,10,100)
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(x,np.tan(x))
plt.title("tan(x)")
```

```python
x=np.linspace(0,10,100)
plt.figure(figsize=(10,5))
plt.subplot(2,2,2)
plt.plot(x,np.exp(-x))
plt.title("exp(-x)")
plt.tight_layout()   #Adjust layout for better spacing
plt.show()
```

```python
cat_col=df.select_dtypes(include="O")
cat_col.head()
```

```python
##features selection --imp step
cat_col=cat_col.drop(["hdd","ssd"],axis=1)
cat_col.head()
```

```python
#encoding --char--num
# 1 lable encoding
#2 -- one hot encoding
pd.get_dummies(cat_col).astype(int).head()
```

```python
num_col=df.select_dtypes(exclude="O")
num_col.head()
```

```python
#concat
final_df=pd.concat([cat_col,num_col],axis=1)
final_df.head()
```

```python
#encoding categorical data
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
ohe=OneHotEncoder(drop="if_binary")
cat_col_encoder=ohe.fit_transform(cat_col).toarray()
cat_col_encoder
```

```python
column_name=list(ohe.get_feature_names_out()) # to get all columns
column_name
```

```python
pd.get_dummies(cat_col,drop_first=True) # it makes only one columns for true or false both
```

```python
one_hot=pd.DataFrame(cat_col_encoder,columns=column_name)
one_hot.head()
```

```python
one_hot=one_hot.reset_index(drop=True)
num_col=num_col.reset_index(drop=True)
final_df=pd.concat([one_hot,num_col],axis=1)
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
x=final_df.drop("increased_price",axis=1)
y=final_df[["increased_price"]]
```

```python
##split data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#model building
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
pred=model.predict(x_test)
y_test["prediction"]=pred
y_test
```

```python
##outliers --remove or transform
q1=df["total_bill"].quantile(0.25)
q3=df["total_bill"].quantile(0.75)
q1
```

```python
q3
```

```python
iqr=q3-q1
iqr
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
lower_bound
```

```python
upper_bound
```

```python
df[df["total_bill"]>upper_bound]
df[df["total_bill"]<lower_bound]
```

```python
df=df[(df["total_bill"]>upper_bound) &
(df["total_bill"]<lower_bound)]
sns.boxplot(df["total_bill"])
```
